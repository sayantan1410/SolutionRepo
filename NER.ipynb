### Main Problem

```python
import pandas as pd
train = pd.read_csv('Airline Booking.csv')
train.head()
```

```python
train.drop(train.iloc[:,2:],inplace = True,axis = 1)
```

```python
train.drop('ID',axis=1)
```

```python
train.dropna(axis = 0,inplace=True)
```

```python
train['Description'][1]
```

```python
raw_string = ""
for ind in train.index:
    raw_string = raw_string + train['Description'][ind] + ". "
```

```python
import spacy

spacy_nlp = spacy.load('en')
document = spacy_nlp(raw_string)
```

```python
for element in document.ents:
    print('Type: %s, Value: %s' % (element.label_, element))
```
